Исходные данные имеют следующую структуру:

````python
#перечень всех документов
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
#перечень полок, на которых хранятся документы (если документ есть в documents, то он обязательно должен быть и в directories)
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}

def cmd_execute(cmd):
    if cmd == 'p':
        num = input('Введите номер документа: ')
        if search(num,1) == None:
            print('Документ не найден в базе')
        else:
            print(f'Владелец документа: {search(num,1)}')
    if cmd == 's':
        num = input('Введите номер документа: ')
        if search(num,2) == None:
            print('Документ не найден в базе')
        else:
            print(f'Документ храниться на полке: {search(num,2)}')
    if cmd == 'l':
        for i in documents:
            print(f"№: {i['number']} тип: {i['type']} Владелец: {i['name']} Полка хранения: {search(i['number'],2)}")       
    if cmd == 'ads':
        num = input('Введите номер полки: ')
        print(oprt_dir(num,1)+' Текущий перечень полок: '+list_dir())
    if cmd == 'ds':
        num = input('Введите номер полки: ')
        print(oprt_dir(num,2)+' Текущий перечень полок: '+list_dir())        
    
def oprt_dir(num,oprt):
    if oprt == 1:
        if num in directories:
            return 'Такая полка уже существует.'
        else:
            directories.setdefault(num,[])
            return 'Полка добавлена.'
    elif oprt == 2:
        if directories.get(num) == None:
            return 'Такой полки не существует.'
        else:
            if directories[num] == []:
                del(directories[num])
                return 'Полка удалена.'
            else:
                return 'На полке есть документа, удалите их перед удалением полки.'
        

def list_dir():
    s = ''
    for i in directories:
        s += i+' '
    return s
    
def search(find,tp):
    if tp == 1:
        for i in documents:
            if i['number'] == find:
                return(i['name'])
    if tp == 2:
        for i in directories:
            if find in directories[i]:
                return i
        
    return 


cmd=''
while cmd != 'q':
    cmd=input('Введите команду: ')
    cmd_execute(cmd)
    
````
