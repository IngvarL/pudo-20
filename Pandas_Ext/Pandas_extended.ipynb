### Задание 1

Для датафрейма log из материалов занятия создайте столбец source_type по следующим правилам:

если источник traffic_source равен yandex или google, то в source_type ставится organic
для источников paid и email из России - ставим ad
для источников paid и email не из России - ставим other
все остальные варианты берем из traffic_source без изменений


```python
import pandas as pd

log = pd.read_csv(r'E:\Pyda-20\Pandas_Ext\visit_log.csv',sep=';')
```

```python
log.head()
```

```python
#Решение с помощью функции с передачей строки
def sourse_type(row_df):
    org = ['yandex','google']
    ad = ['paid','email']    
    if row_df.traffic_source in org:
        return 'organic'
    if (row_df.traffic_source in ad) and (row_df.region in 'Russia'):
        return 'ad'
    if (row_df.traffic_source in ad) and (row_df.region not in 'Russia'):
        return 'other'
    return row_df.traffic_source
```

```python
log['source_type'] = log.apply(sourse_type,axis=1)
```

```python
log.head(15)
```

```python
#Фильрами loc
#Первое условие
org = ['yandex','google']
for i in org:
    log.loc[log.traffic_source.str.contains(i),'source_type'] = 'organic'
#Второе условие
    ad = ['paid','email']
for i in ad:    
    log.loc[(log.traffic_source.str.contains(i)) & (log.region == 'Russia'),'source_type'] = 'ad'
for i in ad:
    log.loc[(log.traffic_source.str.contains(i)) & (log.region != 'Russia'),'source_type'] = 'other'
# Остальные
log.loc[log.source_type is None:,'source_type'] = log.traffic_source
```

### Задание 3

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.


```python
ratings = pd.read_csv(r'E:\Pyda-20\Pandas_Func\ml-latest-small\ratings.csv')
ratings.head()
```

```python
rat100 = ratings.groupby(by=['userId'])['timestamp'].agg(['count','min','max']).reset_index()
rat100.head()
```

```python
rat100['Lifetime'] = rat100['max']-rat100['min']
```

```python
print(f"Среднее время жизни пользователей поставивших более ста оценок фильма — {rat100[rat100['count']>100].Lifetime.mean()}")
```
